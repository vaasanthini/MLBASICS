# Basic Line Plot
##o Plot a simple line graph of 𝑦 = 𝑥^2
##for 𝑥 values ranging from -10 to 10.
##o Label the x-axis as "X Values" and the y-axis as "Squared Values."
##o Add a title to the graph.

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(-10, 10, 400)
y = x**2
plt.plot(x, y)
plt.xlabel("X Values-->")
plt.ylabel("Squared Values-->")
plt.title("Plot of y = x^2")
plt.show()
```

#Bar Plot
##o Create a bar chart showing the marks scored by 5 students: [Sai, 'Vamsi',
##'Divij', 'Sritan', 'Sita'] with marks [85, 90, 78, 92, 88].
##o Add labels to the x-axis and y-axis and give the chart a title.

```python
import matplotlib.pyplot as plt
import numpy as np
students = ['Sai', 'Vamsi', 'Divij', 'Sritan', 'Sita']
marks = [85, 90, 78, 92, 88]
plt.bar(students, marks, color='skyblue')
plt.xlabel("Students-->")
plt.ylabel("Marks-->")
plt.title("Marks Scored by Students")
plt.show()
```

# Scatter Plot
## o Generate two lists of 20 random integers each (for x and y coordinates).
## o Create a scatter plot with these values, and customize it by adding color,
## size, and labels to the points.


```python
import matplotlib.pyplot as plt
import numpy as np
x = np.random.randint(0, 100, 20)
y = np.random.randint(0, 100, 20)
plt.scatter(x, y, c='purple', s=50, marker='o', label='Data points')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Scatter Plot")
plt.legend()
plt.show()
```

#Histogram
## o Simulate 100 random values from a normal distribution using
## np.random.randn() and plot a histogram.
## o Customize the number of bins and add a grid to the plot.

```python
import matplotlib.pyplot as plt
import numpy as np
data = np.random.randn(100)
plt.hist(data, bins=20, color='purple', edgecolor='black')
plt.xlabel("Value-->")
plt.ylabel("Frequency-->")
plt.title("HISTOGRAM OF RANDOM VALUES")
plt.grid(True) #plot a grid
plt.show()  # Display
```

# Pie Chart
##o Visualize the percentage distribution of five categories ['A', 'B', 'C', 'D', 'E']
##with values [20, 30, 25, 15, 10].
##o Display the percentage values on the chart and use a color palette for
##differentiation.

```python
import matplotlib.pyplot as plt
categories = ['A', 'B', 'C', 'D', 'E']
values = [20, 30, 25, 15, 10]
colors = ['orange', 'lightgreen', 'skyblue', 'lightyellow', 'purple']
plt.pie(values, labels=categories, colors=colors, autopct='%1.1f%%', startangle=90)
plt.title("PERCENTAGE DISTRIBUTION OF CATEGORIES")
plt.show()
```

#. Box Plot (Seaborn)
## o Create a box plot to visualize the distribution of the marks scored by
## students in a class: [78, 85, 92, 88, 76, 95, 89, 84, 91, 73].
## o Highlight the median and any outliers.


```python
import seaborn as sns
import matplotlib.pyplot as plt
marks = [78, 85, 92, 88, 76, 95, 89, 84, 91, 73]
sns.boxplot(y=marks, showmeans=True, color='purple') # showmeans = highlight median
plt.ylabel("Marks Scored-->")
plt.title("BOX PLOT OF STUDENT MARKS")
plt.show()
```

# Heatmap (Seaborn)
## o Create a 5x5 matrix with random integers between 1 and 10 using
## np.random.randint() and visualize it as a heatmap using Seaborn.
## o Add a color bar and annotate the heatmap with the values.

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
# Create a 5x5 matrix
matrix = np.random.randint(1, 11, size=(5, 5))
# Create the heatmap
plt.figure(figsize=(8, 6))  # Adjust figure size
sns.heatmap(matrix, annot=True, cmap='viridis', cbar=True) # annot = True to display values
plt.title('Heatmap of a 5x5 Matrix')
plt.show()
```

# 8. Comparison of Matplotlib and Seaborn
## o Generate a pair of plots (bar charts or line plots) using the same data in
## both Matplotlib and Seaborn.
## o Highlight the differences in the code, syntax, and aesthetics of the two
## libraries

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
categories = ['A', 'B', 'C', 'D', 'E'] # Sample data
values = [20, 30, 25, 15, 10]
# Matplotlib bar chart
plt.figure(figsize=(8, 6))  # Adjust figure size
plt.bar(categories, values, color='purple')
plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Bar Chart (Matplotlib)")
plt.show()
# Seaborn bar plot
plt.figure(figsize=(8, 6))
sns.barplot(x=categories, y=values, color = 'black')
plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Bar Plot (Seaborn)")
plt.show()
```

# 9. Pair Plot (Seaborn)
## o Manually create a small dataset with columns such as 'Age', 'Height',
## 'Weight' for 10 individuals and visualize the relationships using a pair plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
# Create a DataFrame
data = {'Age': [25, 30, 22, 28, 35, 27, 29, 31, 24, 26],
        'Height': [172, 173, 160, 170, 182, 188, 179, 172, 181, 186],
        'Weight': [65, 70, 60, 75, 72, 68, 78, 80, 63, 69]}
df = pd.DataFrame(data)
# Create the pair plot
sns.pairplot(df)
plt.title('Pair Plot of Age, Height, and Weight', y=1.02) #add title
plt.show()
```

# 10. Save and Style Plots
## • Plot any graph of your choice and save it in both .png and .pdf formats.
## • Apply at least one built-in style (e.g., 'ggplot' or 'seaborn-darkgrid') and explain
## how the style impacts the visualization.

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0, 10, 100)  # Sample data
y = np.sin(x)
plt.style.use('ggplot')   # Apply 'ggplot' style
# Create the plot
plt.figure(figsize=(6, 6))
plt.plot(x, y, label='sin(x)')
plt.xlabel('x-->')
plt.ylabel('sin(x)-->')
plt.title('Sine Wave Plot')
plt.legend()
plt.grid(True)
# Save the plot in PNG and PDF formats
plt.savefig('sine_plot.png')
plt.savefig('sine_plot.pdf')
# Display the plot
plt.show()
```
